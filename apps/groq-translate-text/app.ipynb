```python
import os
from dotenv import load_dotenv
load_dotenv()

## Configure GROQ API key
groq_api_key = os.getenv("GROQ_API_KEY")

```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(api_key=groq_api_key, model="llama-3.1-8b-instant")

# response = llm.chat(
#     "What is the capital of France?"
# )
# print(response)
```

```python
from langchain_core.messages import SystemMessage, HumanMessage

messages = [
    SystemMessage(content="Translate the following English text to French."),
    HumanMessage(content="Hello How are you?"),
]

result = llm.invoke(messages)
result
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
output_parser.invoke(result)
```

```python
# Using LCEL - chain the components
chain = llm | output_parser
chain.invoke(messages)
```

```python
from langchain_core.prompts import ChatPromptTemplate

generic_template = "Translate the following English text to {language}:"

prompt = ChatPromptTemplate.from_messages(
    [("system",generic_template),("user","{text}")]
)

result = prompt.invoke({"language":"French","text":"Hello How are you?"})
result.to_messages()
```

```python
# Using LCEL
chain = prompt | llm | output_parser
chain.invoke({"language":"German","text":"Hello How are you?"})
```
