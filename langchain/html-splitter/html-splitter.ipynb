```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<html>
<head>
    <title>Example HTML</title>
</head>
<body>
    <h1>Welcome to the Example HTML</h1>
    <p>This is a sample paragraph to demonstrate HTML text splitting.</p>
    <p>It contains multiple sentences and should be split correctly.</p>
</body>
</html>
"""

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2")
]
html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)

html_header_splits = html_splitter.split_text(html_string)
print(html_header_splits)
```

```python
url="https://plato.stanford.edu/entries/goedel/"
headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
    ("h4", "Header 4")
]
html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits = html_splitter.split_text_from_url(url)
html_header_splits
```
